## 1. Load and Inspect the Dataset
We use pandas to read and inspect the Playbook CSV.  
> **Instructor Note**: Stress `.info()` and `.head()` as quick data assessment tools.


```python
import pandas as pd
import matplotlib.pyplot as plt

csv_path = "nist_ai_rmf_playbook.csv"

df = pd.read_csv(csv_path)
print("Shape:", df.shape)
display(df.head(10))
df.info()
```

## 2. Explore Column Names
Columns often begin with function identifiers like `GOVERN`, `MAP`, `MEASURE`, or `MANAGE`.

> **Instructor Note**: Ask students to predict which function dominates.


```python
cols = list(df.columns)
for i, c in enumerate(cols, 1):
    print(f"{i:>3}. {c}")
```

## 3. Classify and Count by Core Function
We count how many columns belong to each of the four AI RMF functions.


```python
core_functions = {'GOVERN': 0, 'MAP': 0, 'MEASURE': 0, 'MANAGE': 0}

for col in df.columns:
    for key in core_functions:
        if str(col).upper().startswith(key):
            core_functions[key] += 1

core_functions
```

## 4. Visualize Distribution
Visualize how many fields correspond to each function.


```python
plt.figure()
plt.bar(core_functions.keys(), core_functions.values())
plt.title("NIST AI RMF — Core Function Distribution (by column prefix)")
plt.xlabel("Core Function")
plt.ylabel("Count of Columns")
plt.tight_layout()
plt.show()
```

## 5. Generate a Manager-Friendly Summary
Outputs both a CSV and a text summary to Colab’s working directory.  
> **Instructor Note**: Have students rewrite summaries for technical and executive audiences.


```python
summary_df = pd.DataFrame(list(core_functions.items()), columns=["Function", "Count"])
summary_csv_path = "rmf_function_summary.csv"
summary_txt_path = "rmf_function_summary.txt"

summary_df.to_csv(summary_csv_path, index=False)

def generate_summary(insights):
    lines = ["AI RMF Core Function Summary for Risk Managers:", ""]
    mapping = {
        "GOVERN": "Policy alignment, accountability, and governance culture.",
        "MAP": "Context mapping, stakeholder identification, and scope definition.",
        "MEASURE": "Evaluation of trustworthiness, bias, performance, and metrics.",
        "MANAGE": "Risk prioritization, mitigation, and lifecycle monitoring."
    }
    for fn, cnt in insights.items():
        lines.append(f"- {fn}: {cnt} items — {mapping.get(fn, '')}")
    return "\n".join(lines)

qual = generate_summary(core_functions)
print(qual)

with open(summary_txt_path, "w", encoding="utf-8") as f:
    f.write(qual)

print(f"Saved: {summary_csv_path}")
print(f"Saved: {summary_txt_path}")
```
